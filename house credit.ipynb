```python
# keep which STATUS is To binary and K
# K is to regular dummy
# To binary to make the exist on 1 and 0 for na
```

```python
variables = pd.read_excel('home credit variables.xlsx')
variables = variables[(variables['file'] == 'static')]
variables = variables[['Variable', 'Variable Type', 'STATUS','Fillna' ,'file']]


def data_process(variables,df_org):

    binary = variables[variables['STATUS']=='To binary']
    dummies = variables[variables['STATUS']=='K']
    delete = variables[variables['STATUS']=='D']

    df_org = df_org.drop(columns=list(delete['Variable']))

    dummies_data = df_org[list(dummies['Variable'])].copy()

    dummies_data = pd.get_dummies(dummies_data, dtype=float)

    df_org = df_org.drop(columns=list(dummies['Variable']))

    binary_data = df_org[list(binary['Variable'])].copy()

    for i in binary_data.columns:
        binary_data.loc[pd.isnull(binary_data[i]),i] = 0
        binary_data.loc[pd.isnull(binary_data[i])==False,i] = 0

    df_org = df_org.drop(columns=list(binary['Variable']))

    df = pd.concat([df_org,binary_data,dummies_data],axis = 1)
    
    # fillna
    fillmissing = variables[variables['Variable Type'] == 'N']
    
    
    for i in range(0,len(fillmissing)):
        fill = fillmissing.iloc[i]['Fillna']
        v = fillmissing.iloc[i]['Variable']
        if(fill == 0):
            df[v].fillna(0,inplace = True)
        elif(fill == 'avg'):
            df[v].fillna(df[v].mean(),inplace = True)
        
    
    
    
    return df
```

# Deposit

```python
# Deposit

deposit = pd.read_csv(path+f'train_deposit_1.csv')


deposit['amount_416A'] = deposit['amount_416A'].fillna(0)

# mark the potential error
deposit['Flag'] = 0

# end date older than startdate
deposit.loc[deposit['contractenddate_991D'] < deposit['openingdate_313D'],'Flag'] = 1

deposit['record_changes_deposit'] = 1

deposit.loc[pd.isnull(deposit['contractenddate_991D']) == False,'contractenddate_991D'] = 1

deposit.loc[pd.isnull(deposit['contractenddate_991D']),'contractenddate_991D'] = 0

tmp = deposit.groupby(['case_id']).agg({'num_group1':'max','record_changes_deposit':'sum'}).reset_index()


deposit.drop(columns = ['record_changes_deposit'],inplace = True)

deposit = deposit.merge(tmp, on = ['case_id','num_group1'],how = 'right')

deposit.drop(columns = ['openingdate_313D'],inplace = True)

deposit.to_excel(f'tmp/deposit_cleaned.xlsx',index = False)
```
